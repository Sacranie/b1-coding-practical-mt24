# Simulate closed-loop
After implementing your control functionality, you can simulate the closed-loop with code that looks something like this...

```python
import os
import sys

# Construct the absolute path to the 'data' directory
project_root = os.path.abspath(os.path.join(os.getcwd(), '..'))
data_dir = os.path.join(project_root, 'data')
csv_file = os.path.join(data_dir, 'mission.csv')

# Add the project root directory to the PYTHONPATH
if project_root not in sys.path:
    sys.path.insert(0, project_root)

# Import necessary classes from the uuv module
import pandas as pd  # Corrected import statement
import numpy as np
import matplotlib.pyplot as plt

from uuv_mission.dynamic import Submarine, ClosedLoop, Mission  # Import directly since in the same package
from uuv_mission.control import Control  # Import Controller from the control module

# Initialize the Submarine instance
sub = Submarine()

# Initialize the Controller with default parameters (previous_error, current_error)
controller_instance = Control(Kd=0.15, Kp= 0.6)

# Create an instance of the ClosedLoop with the Submarine and Controller
closed_loop = ClosedLoop(sub, controller_instance)

# Load mission data from the CSV file
mission = Mission.from_csv(csv_file)

# Run the simulation with random disturbances
trajectory = closed_loop.simulate_with_random_disturbances(mission)

# Plot the completed mission trajectory
trajectory.plot_completed_mission(mission)



```
